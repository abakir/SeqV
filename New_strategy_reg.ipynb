```python
df_ma['7_14_cross'] = ''
for i in range(0,len(df_ma)-1):
  if (df_ma['7_14'][i] * df_ma['7_14'][i+1]) < 0 and df_ma['7_14'][i] > df_ma['7_14'][i+1]:
    df_ma['7_14_cross'][i+2] = 'Short'
  elif (df_ma['7_14'][i]* df_ma['7_14'][i+1]) < 0 and df_ma['7_14'][i] < df_ma['7_14'][i+1]:
    df_ma['7_14_cross'][i+2] = 'Long'
  else:
    df_ma['7_14_cross'][i+2] = 'NA'
```

```python
df_ma['7_28_cross'] = ''
for i in range(0,len(df_ma)-1):
  if (df_ma['7_28'][i] * df_ma['7_28'][i+1]) < 0 and df_ma['7_28'][i] > df_ma['7_28'][i+1]:
    df_ma['7_28_cross'][i+2] = 'Short'
  elif (df_ma['7_28'][i]* df_ma['7_28'][i+1]) < 0 and df_ma['7_28'][i] < df_ma['7_28'][i+1]:
    df_ma['7_28_cross'][i+2] = 'Long'
  else:
    df_ma['7_28_cross'][i+2] = 'NA'
```

```python
df_ma['Mode_1'] = ''
for i in range(0,len(df_ma)):
  if df_ma['7_14'][i] > 0:
    df_ma['Mode_1'][i] = 'Long'
  elif df_ma['7_14'][i] < 0:
    df_ma['Mode_1'][i] = 'Short'
```

```python
df_ma['Mode_2'] = ''
for i in range(0,len(df_ma)):
  if df_ma['7_28'][i] > 0:
    df_ma['Mode_2'][i] = 'Long'
  elif df_ma['7_28'][i] < 0:
    df_ma['Mode_2'][i] = 'Short'
```

```python
# 7 cross 14 then 28
df_ma['Pos_1'] = ''
for i in range(0,len(df_ma)):
  if df_ma['Mode_1'][i] == 'Long' and df_ma['7_28_cross'][i] == 'Long':
      df_ma['Pos_1'][i] = 'Long'
  elif df_ma['Mode_1'][i] == 'Short' and df_ma['7_28_cross'][i] == 'Short':
      df_ma['Pos_1'][i] = 'Short'
  else:
      df_ma['Pos_1'][i] = 'NA'  
```
